```python
with open('input_14.txt') as input_file:
  lines = [[tuple([int(value) for value in point.strip().split(',')]) for point in line.strip().replace('>', '').split('-')] for line in input_file]

def display_cavern(structure):
  x_values = [point[0] for point in structure]
  y_values = [point[1] for point in structure]

  furthest_left = min(x_values)
  furthest_right = max(x_values)
  furthest_up = min(y_values)
  furthest_down = max(y_values)

  output = ''
  for y in range(furthest_up, furthest_down + 1):
    output_line = ''
    for x in range(furthest_left, furthest_right + 1):
      if (x, y) in structure:
        output_line += structure[(x, y)]
      else:
        output_line += '.'
    output_line += '\n'
    output += output_line
  print(output)
```

```python
def draw_structure(lines):
  structure = {(500, 0): '+'}

  for line in lines:
    for i in range(len(line)-1):
      start = line[i]
      end = line[i + 1]
      line_cursor = start
      while line_cursor != end:
        structure[line_cursor] = '#'
        x_delta = end[0] - line_cursor[0]
        y_delta = end[1] - line_cursor[1]
        line_cursor = (line_cursor[0] + (abs(x_delta)//x_delta if x_delta != 0 else 0), line_cursor[1] + (abs(y_delta)//y_delta if y_delta != 0 else 0))
      structure[end] = '#'

  return structure



```

```python

def spawn_sand(structure, min_x, max_x, min_y, max_y, sand_pos=(500, 0)):
  while True:
    if sand_pos[0] < min_x or sand_pos[1] > max_x or sand_pos[1] < min_y or sand_pos[1] > max_y:
      return False
    if (sand_pos[0], sand_pos[1] + 1) not in structure:
      sand_pos = (sand_pos[0], sand_pos[1] + 1)
    elif (sand_pos[0]-1, sand_pos[1] + 1) not in structure:
      sand_pos = (sand_pos[0]-1, sand_pos[1] + 1)
    elif (sand_pos[0]+1, sand_pos[1] + 1) not in structure:
      sand_pos = (sand_pos[0]+1, sand_pos[1] + 1)
    else:
      structure[sand_pos] = 'o'
      return True


def spawn_sand_part_two(structure, sand_pos=(500, 0)):
  while True:
    if (sand_pos[0], sand_pos[1] + 1) not in structure:
      sand_pos = (sand_pos[0], sand_pos[1] + 1)
    elif (sand_pos[0]-1, sand_pos[1] + 1) not in structure:
      sand_pos = (sand_pos[0]-1, sand_pos[1] + 1)
    elif (sand_pos[0]+1, sand_pos[1] + 1) not in structure:
      sand_pos = (sand_pos[0]+1, sand_pos[1] + 1)
    else:
      structure[sand_pos] = 'o'
      if sand_pos == (500, 0):
        return False
      return True

```

```python
structure = draw_structure(lines)
x_values = [point[0] for point in structure]
y_values = [point[1] for point in structure]

furthest_left = min(x_values)
furthest_right = min(y_values)

display_cavern(structure)
for i in range(22):
  spawn_sand(structure, min(x_values), max(x_values), min(y_values), max(y_values))
display_cavern(structure)
```

```python
structure = draw_structure(lines)
x_values = [point[0] for point in structure]
y_values = [point[1] for point in structure]

total_sand = 0
while True:
  if not spawn_sand(structure, min(x_values), max(x_values), min(y_values), max(y_values)):
    break
  total_sand += 1
print(total_sand)
display_cavern(structure)
```

```python
structure = draw_structure(lines)
x_values = [point[0] for point in structure]
y_values = [point[1] for point in structure]

lines.append([(-1000, max(y_values) + 2), (1000, max(y_values) + 2)])
structure = draw_structure(lines)

total_sand = 0
while True:
  if not spawn_sand_part_two(structure):
    total_sand += 1
    break
  total_sand += 1
print(total_sand)
display_cavern(structure)
```
