```python
from collections import deque

with open('input_11.txt') as input_file:
  full_input = input_file.readlines()
while '\n' in full_input:
  full_input.remove('\n')

def interpret_operators(operators, item):
  if operators[0] == 'old':
    op1 = item
  else:
    op1 = int(operators[0])

  if operators[-1] == 'old':
    op2 = item
  else:
    op2 = int(operators[-1])

  if operators[1] == '*':
    return op1*op2
  else:
    return op1+op2

def print_monkey(monkeys):
  for i in range(len(monkeys)):
    print(f'Monkey {i}: {monkeys[i]["items"]}')
```

```python
def parse_input(full_input):
  monkeys = {}
  current_monkey = -1
  for counter, line in enumerate(full_input):
    if counter%6 == 0:
      monkey_name = int(line.strip().split(' ')[1][:-1])
      monkeys[monkey_name] = {}
      current_monkey = monkey_name
    if counter%6 == 1:
      items = [int(x) for x in line.strip().replace(',', '').split(' ')[2:]]
      monkeys[current_monkey]['items'] = deque(items)
    if counter%6 == 2:
      operators = line.strip().split(' ')[-3:]
      monkeys[current_monkey]['operators'] = operators
    if counter%6 == 3:
      test = int(line.strip().split(' ')[-1])
      monkeys[current_monkey]['test'] = test
    if counter%6 == 4:
      true_value = int(line.strip().split(' ')[-1])
      monkeys[current_monkey]['true'] = true_value
    if counter%6 == 5:
      false_value = int(line.strip().split(' ')[-1])
      monkeys[current_monkey]['false'] = false_value
  return monkeys

print_monkey(monkeys)
```

```python
monkeys = parse_input(full_input)
monkey_items = [0 for _ in range(len(monkeys))]

for _ in range(20):
  for i in range(len(monkeys)):
    while monkeys[i]['items']:
      current_item = monkeys[i]['items'].popleft()
      monkey_items[i] += 1
      current_item = interpret_operators(monkeys[i]['operators'], current_item)
      current_item = current_item//3

      if current_item % monkeys[i]['test'] == 0:
        monkeys[monkeys[i]['true']]['items'].append(current_item)
      else:
        monkeys[monkeys[i]['false']]['items'].append(current_item)
monkey_items.sort()
print(monkey_items[-1]*monkey_items[-2])
```
